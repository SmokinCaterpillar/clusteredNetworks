
BRIAN
=====

[BRIAN](http://briansimulator.org/) is an easy-to-use network simulator that lets you define arbitrary neuron models simply by writing down differential equations as strings.

Our Model
---------

We will simulate a network of 2500 randomly connected neurons obeying the following equations:

$$\begin{align}
C\frac{dV}{dt} = - I_L - I_{syn,e} - I_{syn,i} + I_{drive}
\end{align}$$

the differential equation describing the membrane potential development with

$$\begin{align}
I_L = g_L * (V - E_L)
\end{align}$$

the leakage current,

and

$$\begin{align}
I_{syn,x} = g_{syn,x} * (V- E_{syn,x})
\end{align}$$

the synaptic current, where $x \in \{e,i\}$, i.e. whether the synapse is excitatory or inhibitory.

$g_{syn,x}$ will also be a dynamical variable that can evolve over time:

$$\begin{align}
\frac{d g_{syn,x}}{dt}= - \frac{1}{\tau_{syn,x}} g_{syn,x} + \bar{g}_{syn,x} \sum_{t_s \in \text{Spikes}} \delta(t-t_s)
\end{align}$$

Accordingly, the synaptic conductance decays exponentially to 0 but is *kicked* by a factor of $\bar{g}_{syn,x}$ in case of a presynaptic spike.

We will choose $I_{drive}$ such that our network shows constant activity and cells are driven towards the spiking threshold. In order to have some variability in our model, the specific drive of each neuron will be drawn from a random distribution. Yet, we will keep each neuron's driving current constant over time (some people would call this setup *frozen noise*).

Finally, our model has a threshold $V_{th}$. In case a neuron reaches the threshold its membrane voltage is set back to $V_{reset}$. BRIAN will internally do something like the following:

    if V >= V_th:
        V = V_reset

This event counts as a spike which will effect the synaptic currents in other target neurons in the network.


Imports

```python
# Import all stuff we need
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import matplotlib.pylab as pylab

from brian import Network, Equations, NeuronGroup, Connection,\
    SpikeMonitor, raster_plot, StateMonitor, clear, reinit
from brian.stdunits import ms, mV, nS, pA, pF

%matplotlib inline
pylab.rcParams['figure.figsize'] = 12, 8  # changes figure size (width, height) for larger images

clear(True, True) 
reinit()# To reinit brain clocks and remove all old brian objects from namespace,
# it's usually a good idea to put this at the beginning of a script
```

Equation Strings
-----------------

You can specify your model as a string containing the differential equations. So the mathematical formulation from above can be easily translated into a python string:

```python
# The equations defining our neuron model
eqs_string = ''' dV/dt = 1.0/C * (-I_L - I_syn_e - I_syn_i + I_drive) : volt
             I_L = g_L * (V-E_L) : amp
             I_syn_e = g_syn_e * (V-E_syn_e) : amp
             I_syn_i = g_syn_i * (V-E_syn_i) : amp
             dg_syn_e/dt = -1.0/tau_syn_e * g_syn_e : siemens
             dg_syn_i/dt = -1.0/tau_syn_i * g_syn_i : siemens
             I_drive : amp
             '''
```

An equation string contains 4 types of items:

* Dynamical variables: `dx/dt= ... : unit`
    
    They evolve over time and are integrated by BRIAN.
    In our model this is `V` and the synaptic conductances `g_syn_x`.
    
    
* Constant variables: `... = y + ...` 

    These are filled with values if the equations string is turned into an equations object (see below).
    In our model this is for instance the capicatance `C`. Variabels are fixed from the start and are
    equal among all neurons based on the same equation object.
    
    
* Aliases:

  ` ... = z + ...`
  
  ` z = x + y - ...`
  
  These are some filler variables to structure your equations. In our model this is for example `I_L`.
  
  
* Parameters: `k : unit`

  These can be later on set by the user. For instance, to specify individual values for specific neuron subgroups.
  Parameters can potentially change over time by some user introduced actions. 
  For instance, one could think of a sinusoidal input current. 
  Our only parameter is also an input current called `I_drive`.
  Yet, in our case it won't change over time, but will be unique for each individual neuron. 
  
  
Maybe you have noticed that all values have *units*. BRIAN will check for you that all your units match and you don't have to fiddle with the correct orders of magnitude. If your units don't match BRIAN will throw an error.
    


Our Model Parameters
--------------------

```python
# Our model parameters
C = 1200*pF # capacitance of membrane
E_L = -80*mV # Leakage potential
g_L = 5*nS # leakage conductance
g_syn_e_max = 10*nS # increase in exc synaptic conductance when pre-synaptic spike
g_syn_i_max = 50*nS # increase in inh synaptic conductance when pre-synaptic spike
E_syn_e = 0*mV # exc synaptic potential
E_syn_i = -100*mV # inh synaptic potential
tau_syn_e = 5*ms # exc conductance time constant
tau_syn_i = 10*ms # inh conductance time constant
syn_delay = 5*ms # synaptic delay between two neurons
V_th = -50*mV # firing threshold
V_reset = -80*mV # reset potential
refr_period = 2*ms # absolute refractory period

# Our parameters for driving the network to produce activity
I_drive_e_mean = 145*pA # Average drive of exc neuron
I_drive_e_interval = 0.1*I_drive_e_mean # Uniform spread of drive around average
I_drive_i_mean = 140*pA # Average drive of inh neuron
I_drive_i_interval = 0.2*I_drive_i_mean # Uniform spread of drive

# Number of neurons
N_e = 2000 # number of exc neurons
N_i = 500 # number of inh neurons

# Connection probability
conn_prob = 0.05

# Duration of our simulation
duration = 2500*ms
```

Equation Objects
-----------------

Now we turn our string into an equations object.
We will fill all constant variables in the string with the appropriate values.

You can create equation obects via 

`Equations(eqs_string, name_of_variable1_in_eqs=value1, name_of_variable12_in_eqs=value2, ...)`

```python
# Let's create an equation object from our string and parameters
model_eqs = Equations(eqs_string,
                      C=C,
                      g_L=g_L,
                      E_L = E_L,
                      E_syn_e=E_syn_e,
                      E_syn_i=E_syn_i,
                      tau_syn_e=tau_syn_e,
                      tau_syn_i=tau_syn_i)
```

Neuron Groups
--------------

We need to define an assembly of cells following our equations.
We use a so called `NeuronGroup` and pass the following parameters:

* N: Number of neurons
* model: Our equations object
* threshold: A firing threshold
* reset: A reset value
* refractory: Absolute refractory period for which voltage is clamped to reset value

Brian will assume that the first dynamical variable called `V`, `v`, `Vm`, or `vm` in our equation object is the one that is to be reset in case of threshold crossing.

We additionally pass:

* freeze: 

    All equations parameters are replaced by their current values in case the network is started. Accordingly, parameters
    cannot change over time anymore. Yet, this will speed up the simulation. In our case the driving current is fixed
    over the entire course of our simulation, so we can safely use the freeze option.
    
    
* method:

    How to solve the numerical integration. `'Euler'` is the fastest and simplest method.
    
    
* compile:

    BRIAN will try to precompile the equation solver for a speed-up.


Btw, for documentation about equations and neuron groups follow this link:
http://www.briansimulator.org/docs/reference-models-and-groups.html

```python
# Let's create 2500 neurons
all_neurons = NeuronGroup(N=N_e+N_i,
                          model=model_eqs,
                          threshold=V_th,
                          reset=V_reset,
                          refractory=refr_period,
                          freeze = True,
                          method='Euler',
                          compile=True)
```

In principle, you could create separate neuron groups for excitatory and inhibory neurons. In fact, if you have populations of neurons that differ in terms of their model equations, you have to create separate groups. But for us both neuron groups are rather similar and will only differ by their mean drive. Accordingly, we will use subgrouping to divide our population, see below.

Neuron Subgroups
-----------------

You can create subgroups via slicing using the `[`,`]` operator.

We will create two subgroups, 2000 excitatory and 500 inhibitory neurons.

```python
# Divide the neurons into excitatory and inhibitory ones
neurons_e = all_neurons[0:N_e]
neurons_i = all_neurons[N_e:N_e+N_i]
```

Alternatively, you could use the subgroups function instead of slicing:

    neurons_e = all_neurons.subgroup(N_e)
    neurons_i = all_neurons.subgroup(N_i)

Network Drive
--------------

We want to stimulate the network such that it constantly produce activity.
We could choose a constant equal drive for all neurons. Yet, we want some variation in the stimulation.
All neurons should be driven to operate close to the threshold, but some should be driven less and some more. This also means some are driven to constantly fire.

Since we specified `I_drive` in our equations string as a free parameter, wen can choose the parameter for each neuron indivually. If we say `all_neurons.I_drive`, this will return a vector of length `N_e+N_i` containing all driving currents as an array. We can also use the same syntax to set values via `all_neurons.I_drive=...`. 

We will set slighlty different drives for inhibitory and excitatory neurons. The excitatory drive will be on average 145 pA. The individual drives will be chosen randomly from the interval $[145- 0.05*145, 145 + 0.05*145]$ pico ampere. 

Some excitatory cells will receive more than 150 pA and will be driven above the threshold and constantly fire!

The inhibitory neurons will be driven with 140 pA on average and a spread of $0.1*140$ pA. 

Note that NONE of the inhibitory cells will fire due to the driving current! They can only fire if they get excited by the excitatory neurons.

```python
# Divide the neurons into an excitatory and inhibitory subgroup
neurons_e = all_neurons[0:N_e]
neurons_i = all_neurons[N_e:N_e+N_i]

# Set the bias drive, we have slightly different drives for both populations
neurons_e.I_drive = I_drive_e_mean + I_drive_e_interval * np.random.uniform(-0.5, 0.5, N_e)
neurons_i.I_drive = I_drive_i_mean + I_drive_i_interval * np.random.uniform(-0.5, 0.5, N_i)
```

Connections
------------

So far we have 2500 neurons that reside on their own. So we need to connect them to form a network.

There are two ways to form a connection: The new class of [Synapses](http://briansimulator.org/docs/synapses.html) or 
the old-fashioned [Connections](http://briansimulator.org/docs/connections.html). For complicated dynamics and non-linear stuff happening at individual synapses you need to use the former. For simply adding a constant value to your neuron in case of a pre-synaptic spike, you can rely on the more straightforward connection, as we will do.

The general syntacs of a connection is: 

    my_connection = Connection(pre_synaptic_group, post_synaptic_group, variable_that_will_be_increased_by_the_weight,...)

We additionally pass the following arguments:

* delay: Rather self-explanatory

* weight: The value will be added to our target variable `'g_syn_x'` of the target neuron in case of a pre-synaptic spike.

* sparseness: 
    If we specify a weight and sparseness, BRIAN will connect our neurons randomly with the probability we specified here. We choose a very sparse connectivity of $5\%$.

You don't necessarily need random connections, you can create custom connection schemes. Take a look at the documentation if you want to do something like that.

```python
# Set the connections
conn_e = Connection(neurons_e,
                    all_neurons,
                    'g_syn_e',
                    delay=syn_delay,
                    weight = g_syn_e_max,
                    sparseness=conn_prob)

conn_i = Connection(neurons_i,
                    all_neurons,
                    'g_syn_i',
                    delay=syn_delay,
                    weight= g_syn_i_max,
                    sparseness=conn_prob)
```

Our inhibitory and excitatory connections only differ in their target variable and the synaptic weight.

If an excitatory neuron fires an action potential, the excitatory synaptic conductance `'g_syn_e'` (of all target neurons) is increased by the value `g_syn_e_max` (and decays exponentially afterwards). For inhibory neuron activity this is, of course, analogous for the inhibitory synaptic conductance `'g_syn_i'` at the target neurons.

Initial Conditions
-------------------

As initial conditions we will simply choose random membrane potentials and we assume all synaptic conductances to be 0. If you do not initialise dynamical variables, BRIAN will automatically set them to 0.

We can define the initial values of the membrane potentials similar to the driving current from before via `all_neurons.V`. We choose a uniform distribution of initial values from the reset potential to slightly above the threshold. Setting some neurons slightly above the threshold will make them fire at t=0.

```python
# Set the initial membrane potential somewhere between the reversal potential and slightly
# above the firing threshold. This will make some cells fire at t=0.
all_neurons.V = V_reset + np.random.uniform(0, 1.001, N_e+N_i) * (V_th-V_reset)
```

Recording
----------

Before we run our network we need to define so called monitors to record data from our neurons.

We will record all spike times and the membrane potential of a few excitatory and inhibitory neurons.

```python
# Make some monitors to record spikes of all neurons and the membrane potential of a few
spike_mon_e = SpikeMonitor(neurons_e)
spike_mon_i = SpikeMonitor(neurons_i)
state_mon_v_e = StateMonitor(neurons_e, 'V', record=[0,1,2])
state_mon_v_i = StateMonitor(neurons_i, 'V', record=[0,1])
```

The spike monitor is rather self explanatory. 

For the state monitor, besides the neuron group to record from, we need to specify a variable which development we want to record. Here we choose the membrane potential `V`. We could also record a synaptic current like `I_syn_e`, for instance, or any other variable of our equation string. We additionally pass a `record` list to tell BRIAN to only record the variable from the neurons with these particular indices and not from all neurons in the group.

You can find the monitor documentation here: http://briansimulator.org/docs/reference-monitors.html

Running the Network
--------------------

Next, we need to create a network. You can use the `Network` class and simply put all your relevant objects, that is neuron groups (no subgroups!), connections, and monitors into the network. You can pass the elements either to the network constructor or use its `add(...)` method.



```python
# Put everything into the network
network = Network(all_neurons, conn_e, conn_i, spike_mon_e, state_mon_v_e)
# Alternatively use `add`
network.add(spike_mon_i, state_mon_v_i)
```

To run the network for a particular duration use its `run(...)` method. You can pass `report='text'` to make BRIAN print an estimate of the remaining runtime every 10 seconds.

```python
# Let's run our simulation
network.run(duration, report='text')
```

As a side remark, there is the possibility to use BRIAN's so called [run magic function](http://briansimulator.org/docs/reference-network.html#brian.run) instead of constructing a network and using `network.run(...)`. This magic function will search for relevant objects, like neuron groups, in your current namespace. I do not recommend using this functionality. There are two major problems: First, it makes your code incredibly hard to understand. Secondly, if you have to run several networks, potentially due to parameter exploration, using the magic `run` function can create a huge mess. I know that most examples on the BRIAN website use this magic method, but resist the urge to follow these examples! Most of the BRIAN stuff has been coded by scientists. Usually, scientists are not the best programmers in the world :-D

Plotting and Analysis
---------------------

We will plot the network's spiking activity as a raster plot and the evolution of the membrane potential of some neurons.

BRIAN provides it's own `raster_plot` function that directly digests spike monitors: http://briansimulator.org/docs/reference-plotting.html.

By the way, if you need the spike times for further analysis, you cann access these via `spike_mon_e.spikes` as a time ordered list of pairs `(i,t)` where neuron `i` fired at time `t`. Moreover, `spike_mon_e[i]` will return an array of spiketimes of neuron `i`.



```python
# Plot spike raster plots, blue exc neurons, red inh neurons
plt.figure()
gs = gridspec.GridSpec(2, 1, height_ratios=[1, 3])
plt.subplot(gs[0])
raster_plot(spike_mon_i, color='r')
plt.title('Inhibitory neurons')
plt.subplot(gs[1])
raster_plot(spike_mon_e)
plt.title('Excitatory neurons')

# Plot the evolution of the membrane potentials
plt.figure()
for irun in range(1,6):
    plt.subplot(6,1,irun)
    if irun < 3:
        mon = state_mon_v_i
        idx = irun-1
        color='r'
    else:
        mon = state_mon_v_e
        idx = irun-3
        color='b'

    y =  mon.values[idx,:]
    x = mon.times
    plt.plot(x, y, color)
    plt.yticks([min(y), (max(y)-min(y))/2.0+min(y)])

    if irun == 5:
        plt.xlabel('t in ms')
        plt.ylabel('V in volt')

# Show the plots
plt.show()
```

In most cases the network activity will converge to synchronous firing. Synchrony is usually not hard to achieve. Yet, keeping a network in an asynchronous state is a much harder problem (inhibitory STDP is a neat way to do so :-).

Everything at Once
-------------------

Finally, I give you the network script we've gone through one by one as a whole to get a better overview.
Note that the network is rather sensitive to the initial conditions. So if you rerun the whole script, chances are high that your raster and membrane potential plots look different from the previous results.


```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

from brian import Network, Equations, NeuronGroup, Connection,\
    SpikeMonitor, raster_plot, StateMonitor, clear
from brian.stdunits import ms, mV, nS, pA, pF

clear(True, True) 
reinit()# To reinit BRIAN clocks and remove all old BRIAN objects from namespace,
# it's usually a good idea to put this at the beginning of a script

########### Parameters ###############

# The equations defining our neuron model
eqs_string = ''' dV/dt = 1.0/C * (-I_L - I_syn_e - I_syn_i + I_drive) : volt
             I_L = g_L * (V-E_L) : amp
             I_syn_e = g_syn_e * (V-E_syn_e) : amp
             I_syn_i = g_syn_i * (V-E_syn_i) : amp
             dg_syn_e/dt = -1.0/tau_syn_e * g_syn_e : siemens
             dg_syn_i/dt = -1.0/tau_syn_i * g_syn_i : siemens
             I_drive : amp
             '''

# Our model parameters
C = 1200*pF # capacitance of membrane
E_L = -80*mV # Leakage potential
g_L = 5*nS # leakage conductance
g_syn_e_max = 10*nS # increase in exc synaptic conductance when pre-synaptic spike
g_syn_i_max = 50*nS # increase in inh synaptic conductance when pre-synaptic spike
E_syn_e = 0*mV # exc synaptic potential
E_syn_i = -100*mV # inh synaptic potential
tau_syn_e = 5*ms # exc conductance time constant
tau_syn_i = 10*ms # inh conductance time constant
syn_delay = 5*ms # synaptic delay between two neurons
V_th = -50*mV # firing threshold
V_reset = -80*mV # reset potential
refr_period = 2*ms # absolute refractory period

# Our parameters for driving the network to produce activity
I_drive_e_mean = 145*pA # Average drive of exc neuron
I_drive_e_interval = 0.1*I_drive_e_mean # Uniform spread of drive around average
I_drive_i_mean = 140*pA # Average drive of inh neuron
I_drive_i_interval = 0.2*I_drive_i_mean # Uniform spread of drive

# Number of neurons
N_e = 2000 # number of exc neurons
N_i = 500 # number of inh neurons

# Connection probability
conn_prob = 0.05

# Duration of our simulation
duration = 2500*ms


########### Network Creation ############

# Let's create an equation object from our string and parameters
model_eqs = Equations(eqs_string,
                      C=C,
                      g_L=g_L,
                      E_L = E_L,
                      E_syn_e=E_syn_e,
                      E_syn_i=E_syn_i,
                      tau_syn_e=tau_syn_e,
                      tau_syn_i=tau_syn_i)


# Let's create 2500 neurons
all_neurons = NeuronGroup(N=N_e+N_i,
                          model=model_eqs,
                          threshold=V_th,
                          reset=V_reset,
                          refractory=refr_period,
                          freeze = True,
                          method='Euler',
                          compile=True)


# Divide the neurons into an excitatory and inhibitory subgroup
neurons_e = all_neurons[0:N_e]
neurons_i = all_neurons[N_e:N_e+N_i]

# Set the bias drive, we have slightly different drives for both populations
neurons_e.I_drive = I_drive_e_mean + I_drive_e_interval * np.random.uniform(-0.5, 0.5, N_e)
neurons_i.I_drive = I_drive_i_mean + I_drive_i_interval * np.random.uniform(-0.5, 0.5, N_i)


# Set the connections
conn_e = Connection(neurons_e,
                    all_neurons,
                    'g_syn_e',
                    delay=syn_delay,
                    weight = g_syn_e_max,
                    sparseness=conn_prob)

conn_i = Connection(neurons_i,
                    all_neurons,
                    'g_syn_i',
                    delay=syn_delay,
                    weight= g_syn_i_max,
                    sparseness=conn_prob)


# Set the initial membrane potential somewhere between the reversal potential and slightly
# above the firing threshold. This will make some cells fire at t=0.
all_neurons.V = V_reset + np.random.uniform(0, 1.001, N_e+N_i) * (V_th-V_reset)

# Make some monitors to record spikes of all neurons and the membrane potential of a few
spike_mon_e = SpikeMonitor(neurons_e)
spike_mon_i = SpikeMonitor(neurons_i)
state_mon_v_e = StateMonitor(neurons_e, 'V', record=[0,1,2])
state_mon_v_i = StateMonitor(neurons_i, 'V', record=[0,1])

# Put everything into the network
network = Network(all_neurons, conn_e, conn_i, spike_mon_e, state_mon_v_e)
# Alternatively use `add`
network.add(spike_mon_i, state_mon_v_i)


############ Simulation Execution ###############

# Let's run our simulation
network.run(duration, report='text')


############### Plotting ####################

# Plot spike raster plots, blue exc neurons, red inh neurons
plt.figure()
gs = gridspec.GridSpec(2, 1, height_ratios=[1, 3])
plt.subplot(gs[0])
raster_plot(spike_mon_i, color='r')
plt.title('Inhibitory neurons')
plt.subplot(gs[1])
raster_plot(spike_mon_e)
plt.title('Excitatory neurons')

# Plot the evolution of the membrane potentials
plt.figure()
for irun in range(1,6):
    plt.subplot(6,1,irun)
    if irun < 3:
        mon = state_mon_v_i
        idx = irun-1
        color='r'
    else:
        mon = state_mon_v_e
        idx = irun-3
        color='b'

    y =  mon.values[idx,:]
    x = mon.times
    plt.plot(x, y, color)
    plt.yticks([min(y), (max(y)-min(y))/2.0+min(y)])

    if irun == 5:
        plt.xlabel('t in ms')
        plt.ylabel('V in volt')

# Show the plots
plt.show()
```
