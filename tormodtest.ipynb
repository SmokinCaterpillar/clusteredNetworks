```python
# Import all stuff we need
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import matplotlib.pylab as pylab

from brian import Network, Equations, NeuronGroup, Connection,\
    SpikeMonitor, raster_plot, StateMonitor, clear, reinit
from brian.stdunits import ms, mV, nS, pA, pF

%matplotlib inline
pylab.rcParams['figure.figsize'] = 12, 8  # changes figure size (width, height) for larger images

clear(True, True) 
reinit()# To reinit brain clocks and remove all old brian objects from namespace,
# it's usually a good idea to put this at the beginning of a script



# The equations defining our neuron model
eqs_string2 = ''' dV/dt = (1.0/tau)*(myu-v) + Isyne + Isyni : volt
             Isyni = 1/(tau2 - tau1) * IsyniDerFac : amp
             IsyniDerFac = IsyniDerFac2 - IsyniDerFac1
             dIsyniDerFac1/dt = -IsyniDerFac1/tau1
             dIsyniDerFac2/dt = -IsyniDerFac2/tau2
             Isyne = 1/(tau2 - tau1) * IsyneDerFac : amp
             IsyneDerFac = IsyneDerFac2 - IsyneDerFac1
             dIsyneDerFac1/dt = -IsyneDerFac1/tau1
             dIsyneDerFac2/dt = -IsyneDerFac2/tau2
             '''

eqs_string = ''' dV/dt = (1.0/tau)*(myu-v) + Isyn : volt
             Isyn = 1/(tau2 - tau1) * IsynDerFac : Hz
             dIsynDerFac/dt = -IsynDerFac * (tau1*exp(t/tau1) - tau2*exp(t/tau2)) / (tau1*tau2*(exp(t/tau1) - exp(t/tau2))) : Hz
             myu : 1
             '''




# Our model parameters
taue = 15*ms # time constant of membrane excitatory
taui = 10*ms # time constant of membrane inhibitory
tau1 = 1*ms
tau2e = 3*ms
tau2i = 2*ms
myueMax = 1.2
myueMin = 1.1
myuiMax = 1.05
myuiMin = 1.0
syn_delay = 5*ms # synaptic delay between two neurons
V_th = 1*mV # firing threshold
V_reset = 0*mV # reset potential
refr_period = 5*ms # absolute refractory period
bump = 1 #value to bump Isyn with on spike
jEE = 0.024
jEI = -0.045
jIE = 0.014
jII = -0.057
pEI = 0.5
pIE = pEI
pII = pEI
pEE = 0.2

# Our parameters for driving the network to produce activity
I_drive_e_mean = 145*pA # Average drive of exc neuron
I_drive_e_interval = 0.1*I_drive_e_mean # Uniform spread of drive around average
I_drive_i_mean = 140*pA # Average drive of inh neuron
I_drive_i_interval = 0.2*I_drive_i_mean # Uniform spread of drive

# Number of neurons
N_e = 2000 # number of exc neurons
N_i = 500 # number of inh neurons

# Connection probability
conn_prob = 0.05

# Duration of our simulation
duration = 2500*ms




# Let's create an equation object from our string and parameters
model_eqs_e = Equations(eqs_string,
                      tau=taue,
                      tau1=tau1,
                      tau2=tau2e,
                      )

model_eqs_i = Equations(eqs_string,
                      tau=taui,
                      tau1=tau1,
                      tau2=tau2i,
                      )





# Divide the neurons into excitatory and inhibitory ones
neurons_e = NeuronGroup(N=N_e,
                          model=model_eqs_e,
                          threshold=V_th,
                          reset=V_reset,
                          refractory=refr_period,
                          freeze = True,
                          method='Euler',
                          compile=True)

neurons_i = NeuronGroup(N=N_i,
                          model=model_eqs_i,
                          threshold=V_th,
                          reset=V_reset,
                          refractory=refr_period,
                          freeze = True,
                          method='Euler',
                          compile=True)

all_neurons = []
all_neurons.append(neurons_e)
all_neurons.append(neurons_i)


# Set the bias drive, we have slightly different drives for both populations
neurons_e.myu = np.random.uniform(myueMin, myueMax, N_e)
neurons_i.myu = np.random.uniform(myuiMin, myuiMax, N_i)





# Set the connections
conn_ee = Connection(neurons_e,
                    neurons_e,
                    'IsynDerFac',
                    delay=syn_delay,
                    weight = jEE*bump,
                    sparseness=pEE)

conn_ei = Connection(neurons_e,
                    neurons_i,
                    'IsynDerFac',
                    delay=syn_delay,
                    weight = jEI*bump,
                    sparseness=pEI)

conn_ie = Connection(neurons_i,
                    neurons_e,
                    'IsynDerFac',
                    delay=syn_delay,
                    weight = jIE*bump,
                    sparseness=pIE)

conn_ii = Connection(neurons_i,
                    neurons_i,
                    'IsynDerFac',
                    delay=syn_delay,
                    weight = jII*bump,
                    sparseness=pII)



# Set the initial membrane potential somewhere between the reversal potential and slightly
# above the firing threshold. This will make some cells fire at t=0.
all_neurons.V = V_reset + np.random.uniform(0, 1.001, N_e+N_i) * (V_th-V_reset)




# Make some monitors to record spikes of all neurons and the membrane potential of a few
spike_mon_e = SpikeMonitor(neurons_e)
spike_mon_i = SpikeMonitor(neurons_i)
state_mon_v_e = StateMonitor(neurons_e, 'V', record=[0,1,2])
state_mon_v_i = StateMonitor(neurons_i, 'V', record=[0,1])




# Put everything into the network
network = Network(all_neurons, conn_ee, conn_ei, conn_ie, conn_ii, spike_mon_e, state_mon_v_e)
# Alternatively use `add`
network.add(spike_mon_i, state_mon_v_i)




# Let's run our simulation
network.run(duration, report='text')




# Plot spike raster plots, blue exc neurons, red inh neurons
plt.figure()
gs = gridspec.GridSpec(2, 1, height_ratios=[1, 3])
plt.subplot(gs[0])
raster_plot(spike_mon_i, color='r')
plt.title('Inhibitory neurons')
plt.subplot(gs[1])
raster_plot(spike_mon_e)
plt.title('Excitatory neurons')

# Plot the evolution of the membrane potentials
plt.figure()
for irun in range(1,6):
    plt.subplot(6,1,irun)
    if irun < 3:
        mon = state_mon_v_i
        idx = irun-1
        color='r'
    else:
        mon = state_mon_v_e
        idx = irun-3
        color='b'

    y =  mon.values[idx,:]
    x = mon.times
    plt.plot(x, y, color)
    plt.yticks([min(y), (max(y)-min(y))/2.0+min(y)])

    if irun == 5:
        plt.xlabel('t in ms')
        plt.ylabel('V in volt')

# Show the plots
plt.show()
```
